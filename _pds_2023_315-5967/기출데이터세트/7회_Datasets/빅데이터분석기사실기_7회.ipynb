03.	주어진 데이터 세트는 아파트 가격 데이터다. 아파트 가격을 결정하는 10개의 독립변수와 1개의 종속변수가 있다. 각 독립변수의 이상치를 사분위수를 기준으로 측정하고 이상치가 가장 많은 독립변수의 이상치 개수를 구하시오. 이때 결측치를 먼저 제거한 후 이상치를 구하시오. [단, 이상치는 (1Q-1.5IQR)보다 작거나 (3Q+1.5IQR)보다 큰 값이며, IQR은 Q3에서Q1을 뺀 값이다.]

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd

df3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7_1_3_apartment_prices_dataset.csv', encoding='cp949')
df3.head()
```

```python
# 결측치 제거
df3_cleaned = df3.dropna()
df3_cleaned.info()
```

```python
# 각 독립 변수의 이상치 계산
outliers_count = {}
for column in df3_cleaned.columns[:-1]:  # 가격(Price) 제외
    Q1 = df3_cleaned[column].quantile(0.25)
    Q3 = df3_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df3_cleaned[(df3_cleaned[column] < lower_bound) | (df3_cleaned[column] > upper_bound)]
    outliers_count[column] = outliers.shape[0]
```

```python
# 가장 많은 이상치를 가진 변수 탐색
results = max(outliers_count, key=outliers_count.get)
print(results)
```

**2) 작업형 제 2유형**
01.	회사에서 신규 영업지점을 설립하려고 한다. 기존 영업지점 데이터를 이용하여 신규 영업지점을 설립할 경우 달성 가능한 매출을 예측하여 의사결정에 사용할 계획이다. 주어진 훈련 데이터 세트를 활용하여 매출 예측 머신러닝 모델을 구축하고, 테스트 데이터를 대상으로 매출을 예측한 결과를 CSV 파일로 제출하시오.
* 결과 제출 양식 : 예측한 Sales 데이터를 기준으로 RMSE를 계산하여 채점에 활용할 예정


```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import mean_squared_error
```

```python
# 데이터 불러오기
train_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7_2_1_sales_train_dataset.csv')
test_data_x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7_2_2_sales_test_dataset_x.csv')
test_data_y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7_2_3_sales_test_dataset_y.csv')
```

```python
# 데이터 확인
train_data.head()
```

```python
# 변수 구분
# 범주형 칼럼 선택
categorical_columns = ['BranchName', 'City', 'IndustryType']

# 수치형 칼럼 선택
numeric_features = ['Population', 'IncomeGeneratingPopRatio', 'AverageIncome']
```

```python
# 훈련 데이터와 테스트 데이터를 결합하여 원-핫 인코딩
combined = pd.concat([train_data, test_data_x])
combined_encoded = pd.get_dummies(combined, columns=categorical_columns)
```

```python
# 인코딩된 데이터를 다시 훈련 데이터와 테스트 데이터로 분리
train_data_encoded = combined_encoded.iloc[:len(train_data)]
test_data_x_encoded = combined_encoded.iloc[len(train_data):].drop(['Sales'], axis=1)
```

```python
# 수치형 데이터 스케일링
scaler = StandardScaler()
train_data_encoded[numeric_features] = scaler.fit_transform(train_data_encoded[numeric_features])
test_data_x_encoded[numeric_features] = scaler.transform(test_data_x_encoded[numeric_features])
```

```python
# 훈련 데이터와 테스트 데이터 준비
X_train_encoded = train_data_encoded.drop('Sales', axis=1)
y_train = train_data_encoded['Sales']
X_test_encoded = test_data_x_encoded
y_test = test_data_y['Sales']
```

```python
# 선형 회귀 모델 학습 및 평가
linear_model = LinearRegression()
linear_model.fit(X_train_encoded, y_train)
linear_pred = linear_model.predict(X_test_encoded)
linear_rmse = np.sqrt(mean_squared_error(y_test, linear_pred))
print("Linear Regression RMSE:", linear_rmse)
```

```python
# 랜덤 포레스트 모델 학습 및 평가
random_forest_model = RandomForestRegressor(n_estimators=100)
random_forest_model.fit(X_train_encoded, y_train)
rf_pred = random_forest_model.predict(X_test_encoded)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
print("Random Forest RMSE:", rf_rmse)
```

```python
# 인공신경망 모델 학습 및 평가
mlp_model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=500)
mlp_model.fit(X_train_encoded, y_train)
mlp_pred = mlp_model.predict(X_test_encoded)
mlp_rmse = np.sqrt(mean_squared_error(y_test, mlp_pred))
print("MLP Regressor RMSE:", mlp_rmse)
```

```python
# 테스트 데이터에 대한 판매 예측
results = random_forest_model.predict(X_test_encoded)

# 예측 결과를 CSV 파일로 저장
results_df = pd.DataFrame({'PredictedSales': results})
results_df.to_csv('000000.csv', index=False)
```

**3) 작업형 제 3유형**

01.	키(Height), 몸무게(Weight), 발 사이즈(Shoe Size)를 이용하여 남녀(Sex) 성별을 분류하는 로지스틱 회귀모형을 만들고자 한다. 주어진 데이터 세트를 이용하여 로지스틱 회귀모형을 만들고 아래의 질문에 답하시오. 단, 주어진 데이터 세트의 2/3을 훈련용으로, 1/3을 테스트용으로 분할하여 작업하시오


```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, accuracy_score
```

```python
# 데이터 불러오기
df4 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7_3_1_gender_prediction_dataset.csv')
df4.head()
```

```python
# 데이터 분할 (성별을 0과 1로 인코딩)
X = df4[['Height', 'Weight', 'ShoeSize']]
y = (df4['Gender'] == 'Male').astype(int)  # 'Male'을 1로, 'Female'을 0으로 변환
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

(1)	훈련용 데이터를 이용하여 ‘Height’, ‘Weight’, ‘Shoe Size’를 설명변수로 사용하는 로지스틱 회귀모형을 만들고 Weight의 오즈비를 계산하시오.

```python
# (1) 로지스틱 회귀 모델 학습
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
# Weight의 Odds Ratio 계산
results = np.exp(log_reg.coef_[0][1])
print(results)
```

(2)	모델의 로짓우도(log-likelihood)를 구하시오.

```python
results = -log_loss(y_train, log_reg.predict_proba(X_train), normalize=False)
print(results)
```

(3)	(2)에서 만든 모델로 평가용 데이터를 예측한 결과와 실제값의 오차율을 구하시오.

```python
y_pred = log_reg.predict(X_test)
results = 1 - accuracy_score(y_test, y_pred)
print(results)
```

**3) 작업형 제 3유형**

02.	중고차의 가격을 다중회귀분석으로 분석하고자 한다. ‘Price’를 종속변수로, ‘Year’, ‘Transmission’, ‘Mileage’, ‘Fuel type’, ‘Tax’, ‘Mpg’, ‘Engine Size’를 독립변수로 사용한다. 데이터의 구성은 아래 표와 같다.

(1)	다중회귀분석 모형의 결정계수를 구하시오.

```python
import pandas as pd
import statsmodels.api as sm
```

```python
# 데이터 불러오기
df5 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/7_3_2_used_car_price_dataset.csv')
df5.head()
```

```python
# 독립 변수 범주형 데이터 처리 및 상수항 추가
# 독립 변수에 대한 더미 변수 생성 (범주형 데이터 처리)
X = pd.get_dummies(df5.drop('price', axis=1), drop_first=True)
y = df5['price']

# 상수항 추가
X_const = sm.add_constant(X)
```

```python
# 다중 회귀 분석 모델 생성
model = sm.OLS(y, X_const).fit()
```

```python
# 결정 계수(R-squared)
results = model.rsquared
print(results)
```

(2)	독립변수 중 회귀계수가 가장 큰 변수와 그 값을 구하여 튜플 타입으로 출력하시오.
* 예시 : (‘Weight’, 0.09003569454430438)


```python
# 가장 큰 회귀 계수와 그 변수
coefficients = model.params
max_coef_var, max_coef_value = coefficients.idxmax(), coefficients.max()
print((max_coef_var, max_coef_value))
```

(3)	독립변수 중 p-value가 가장 낮은 변수와 그 값을 구하여 튜플 타입으로 출력하시오.

```python
# 가장 낮은 p-value와 그 변수
p_values = model.pvalues
min_pval_var, min_pval_value = p_values.idxmin(), p_values.min()
print((min_pval_var, min_pval_value))
```
