```python
df = pd.read_csv('/Users/s/dev_book/exam_practice/02/data/P2306/P230601.csv')
df 
```

```python
df['call_time'] = pd.to_datetime(df['신고일시'])
df['arrive_time'] = pd.to_datetime(df['출동일시'])
df['station'] = df['출동소방서']

df['diff_time'] = (df['arrive_time'] - df['call_time']).dt.total_seconds()
```

```python
#연도별 월평균, 가장늦게 출동한 출동소방서의 월평균 시간 / 반올림 30초 단위로
df = df.groupby( [df['station'],\
           df['call_time'].dt.year,
           df['call_time'].dt.month]).mean('diff_time')
df
```

```python
df = df.sort_values('diff_time', ascending = False)
df
```

```python
resnum = df['diff_time'].head(1).iloc[0] / 60
resnum
```

```python
import pandas as pd 
from scipy.stats import chisquare
```

```python
df5 = pd.read_csv('/Users/s/dev_book/exam_practice/02/data/P2306/P230605.csv',encoding="euc-kr")
```

```python
dt_placebo = df5.groupby('코드').size().reset_index(name='건수')
```

```python
dt_placebo
```

```python
df5.groupby('코드').count()
```

```python
df5.groupby('코드').size()
```

```python
dt_placebo['비율'] = dt_placebo['건수']/len(df5)
```

```python
dt_placebo
```

```python
round(dt_placebo['비율'][3], 3)
```

```python
dt_rate = pd.DataFrame({'코드':[1,2,3,4] , '비율': [0.05, 0.1, 0.05, 0.8]})
res_chisq = chisquare(dt_placebo['건수'], f_exp = dt_rate['비율']*len(df5))
res_chisq
```

```python
round(res_chisq.statistic, 3)
```

```python
round(res_chisq.pvalue, 3)
```
