```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

# create an instance of graph
G = nx.Graph()

# add nodes 
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)

# add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1,4)

# visualize graph
plt.figure()
nx.draw(G, with_labels=True, node_size=1000, font_size=30)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python

# create adjacency matrix
adj = np.array([[0, 1, 1, 1],
                [1, 0, 0, 0],
                [1, 0, 0, 0],
                [1, 0, 0, 0]])

# graph using adjacency matrix 
G = nx.DiGraph(np.array(adj))

# visualize graph
plt.figure(figsize=(10, 6))
nx.draw(G, with_labels=True, node_size=3000, font_size=40)
```

```python
import networkx as nx
import numpy as np
from itertools import combinations

# adjacency matrix
adj = [[0,1,1,1],[1,0,1,1],[1,1,0,1],[1,1,1,0]]

# create graph instance
G = nx.DiGraph(np.array(adj))
 #nx.DiGraph(np.array(adj))
# check if G is fully connected
def is_fully_connected(G):
    for (i,j) in combinations(G.nodes(),2): 
        if not G.has_edge(i,j):
            print("Graph is NOT fully connected!")
    print("Graph is fully connected!")

is_fully_connected(G)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# generate random embedding matrix
np.random.seed(42)
embedding = np.random.rand(78,16)

scaler = StandardScaler()
embedding = scaler.fit_transform(embedding)

# pca
pca = PCA(n_components = 2)
embedding_pca = pca.fit_transform(embedding)

# plot
plt.figure(figsize=(10, 6))
plt.scatter(embedding_pca[:, 0], embedding_pca[:, 1], s=100)
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# generate random embedding matrix
np.random.seed(42)
embedding = np.random.rand(78,16)

# tsne
tsne = TSNE(n_components=2,random_state=42)
embedding_tsne = tsne.fit_transform(embedding)

# plot
plt.figure(figsize=(10, 6))
plt.scatter(embedding_tsne[:, 0], embedding_tsne[:, 1], s=100)
```
